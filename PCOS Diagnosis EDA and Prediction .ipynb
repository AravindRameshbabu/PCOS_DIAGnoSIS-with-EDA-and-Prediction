```python
#### Handling the imbalance data with Imblearn 
```

```python
import imblearn
print(imblearn.__version__)
```

```python
from imblearn.over_sampling import RandomOverSampler

random = RandomOverSampler()
```

```python
X_Random_over_sample , y_Random_over_sample = random.fit_resample(X,y)
```

```python
print(X.shape)
print(y.shape)

print('*'*50)

print(X_Random_over_sample.shape)
print(y_Random_over_sample.shape)
```

```python
## After 
plt.figure(figsize=(12,10))
y_Random_over_sample.value_counts().plot.pie(autopct='%.2f')
plt.title('Results after handling the imbalance data with imblearn')
plt.show()
```

```python
### Lets split the data 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_Random_over_sample,y_Random_over_sample,test_size=0.2,random_state=42)
```

```python
#### Even though it is classification problem we need to scale the data with standard scalar
```

```python
from sklearn.preprocessing import StandardScaler
Std = StandardScaler()
```

```python
print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

```python
x_train = Std.fit_transform(x_train)
x_test = Std.transform(x_test)
```

```python
print(x_train.shape)
print(x_test.shape)

print('*'*100)

print(y_train.shape)
print(y_test.shape)
```

##### Random Forest with Grid Search  

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Random = RandomForestClassifier(random_state=42)
```

```python
### Defining the parameters 
```

```python
parameter_grids={
    'n_estimators':[100,200,300,400],
    'max_depth':[10,20,30,40,None],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'max_features':['sqrt','log2'],
    'class_weight':['balanced',None]
}
```

```python
### Performing the grid search
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search  = GridSearchCV(
            estimator=Random,
            param_grid=parameter_grids,
            cv = 5,
            scoring='f1',
            n_jobs=-1,
            verbose=2)


grid_search.fit(x_train,y_train)
```

```python
### Prediction part 
```

```python
random_forest_training_prediction = grid_search.best_estimator_.predict(x_train)
random_forest_test_prediction = grid_search.best_estimator_.predict(x_test)
```

##### Evaluting the model's performance with accuracy score, f1score,classification report and confusion matrix

```python
print('Accuracy score')
print(' '*100)
print('*'*100)
print(' '*100)

print('This is the accuracy score of training data with Random forest: ',accuracy_score(random_forest_training_prediction,y_train)*100)
print(' '*100)
print('*'*100)
print(' '*100)
print('This is the accuracy score of test data with Random forest: ',accuracy_score(random_forest_test_prediction,y_test)*100)
print(' '*100)
print('*'*100)
print(' '*100)

print('f1_score')
print('This is the f1 score of training data with random forest: ',f1_score(random_forest_training_prediction,y_train)*100)
print(' '*100)
print('*'*100)
print(' '*100)
print('This is the f1 score of test data with random forest: ',f1_score(random_forest_test_prediction,y_test)*100)

print('Classification Report')
print(' '*100)
print('*'*100)
print(' '*100)

print('This is the classification report on training data with random forest: ',classification_report(random_forest_training_prediction,y_train))
print(' '*100)
print('*'*100)
print(' '*100)
print('This is the classification report on test data with random forest: ',classification_report(random_forest_test_prediction,y_test))
print(' '*100)
print('*'*100)
print(' '*100)

print('Confusion Matrix')
print('This is the confusion matrix of training data with random forest: ',confusion_matrix(random_forest_training_prediction,y_train))
print(' '*100)
print('*'*100)
print(' '*100)
print('This is the confusion matrix of test data with random forest: ',confusion_matrix(random_forest_test_prediction,y_test))
```
